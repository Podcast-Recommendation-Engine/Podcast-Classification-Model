```python
def is_row_kid_friendly(keywords, bad_list):
    if not isinstance(keywords, list):
        return 1  # approprié

    keywords = [kw.strip().lower() for kw in keywords]

    # Vérifier si un élément de la ligne est dans la liste non appropriée
    for kw in keywords:
        if kw in bad_list:
            return 0  # non approprié
    
    return 1  # approprié
df["is_kid_friendly"] = df["keywords_clean"].apply(
    lambda x: is_row_kid_friendly(x, non_kid_keywords)
)
```
